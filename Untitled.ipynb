```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pypylon.pylon as py
```

```python
tlf = py.TlFactory.GetInstance()
mycam = py.InstantCamera(tlf.CreateFirstDevice())
mycam.Open()
mycam.UserSetSelector = "Default"
mycam.UserSetLoad.Execute()
```

```python
mycam.PixelFormat = "Mono10"
mycam.Gain = 10
```

```python
res = mycam.GrabOne(1000)
img = res.Array
plt.imshow(img)
```

```python
mycam.Gain = 0
res2 = mycam.GrabOne(1000)
img2 = res2.Array
plt.imshow(img2)
```

```python
from scipy import ndimage
def get_centroid (ndarray):
    centroid = ndimage.center_of_mass((ndarray))
    return (centroid)
```

```python
center = get_centroid(img2)
```
